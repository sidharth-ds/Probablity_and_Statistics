```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Poisson distribution (discrete):
* focuses on larger distribution area
* lambda value : is like Mean value or expected value.

```python
z=np.random.poisson(lam=60,size=1000)

plt.hist(z,bins=20)
plt.show()
plt.plot(z)
plt.show()
sns.kdeplot(z)
plt.show()
```

* comparison of various distributions:

```python
x=np.random.normal(loc=60,scale=4,size=1000)
y=np.random.binomial(n=100,p=0.6,size=1000)
z=np.random.poisson(lam=60,size=1000)

sns.kdeplot(x,label="normal")
sns.kdeplot(y,label="binomial")
sns.kdeplot(z,label="poisson")

plt.legend()
plt.show()
```

## Uniform distribution:

* no peak at the mean
* data is distributed evenly (even y_axis)
* by default value is : 0 to 1
* to change the range: use - low, high 

```python
m=np.random.uniform(size=1000)

plt.hist(m,bins=20)
plt.show()
plt.plot(m)
plt.show()
sns.kdeplot(m)
plt.show()
```

```python
m=np.random.uniform(low=40,high=80,size=1000)

plt.hist(m,bins=20)
plt.show()
plt.plot(m)
plt.show()
sns.kdeplot(m)
plt.show()
```

* comparison of various distributions:

```python
x= np.random.normal(loc=60,scale=4,size=1000)
y=np.random.binomial(n=100,p=0.6,size=1000)
z=np.random.poisson(lam=60,size=1000)
m=np.random.uniform(low=40,high=80,size=1000)

sns.kdeplot(x,label="normal")
sns.kdeplot(y,label="binomial")
sns.kdeplot(z,label="poisson")
sns.kdeplot(m,label="uniform")
plt.legend()
plt.show()
```

## Chi_Squared test:
* null hypothesis/ alternate hypothesis
* used for mutually exclusive statements
* assumption:(eg: court case)
  - null hypothesis (innocent)
  - alternate hypothesis (guilty)
* type-1 error : ~innocent
* type-2 error : ~guilty
* formula:
    - ((observed val - expected val)**2) / expected value

Application of chi squared test:
* A chi-square test is a statistical test used to compare observed results with expected results. 
* The purpose of this test is to determine if a difference between observed data and expected data is due to chance, or if it is due to a relationship between the variables you are studying.

## Chi aquare distribution: only for Classification data(discrete), not for Regression data(continuous)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

concept: 
* Degree of Freedom : logically indepedent variables. 
  - (total no:of rows-1) multiplied by (total no:of columns-1)
  - total no:of column - 1
  - one column is dependent variable (target variable)
* Threshold Limits : alpha --- significance value.
  - values beyond alpha will be ignored in both direction.

```python
x = np.random.chisquare(df=5,size=1000)

plt.plot(x)
plt.show()

plt.hist(x)
plt.show()

sns.kdeplot(x)
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
df = load_iris()
df
```

splitting the dataset into:
* train data ---- (data)
* test data ---- (target)

```python
x = df.data
x
```

```python
y = df.target     # here, species is taken as target data
y
```

```python
X = x.astype(int)  # to convert float into int
X
```

```python
chi2_features = SelectKBest(chi2,k=3)  # automatically selects the best (3 features)  # SelectKBest works based on chisquare_distribution & k_value
x_ch = chi2_features.fit_transform(X,y)  # X=data(train), y=target(test)
                                        # k is DoF. ie. no:of column features / independent variables
```

```python
x_ch.shape, x.shape
```

```python
chi2_features = SelectKBest(chi2,k=2)  # automatically selects the best(2 features)   
x_ch = chi2_features.fit_transform(X,y)  # X=data(train), y=target(test)
                                        # k is DoF. ie. no:of column features / independent variables
```
