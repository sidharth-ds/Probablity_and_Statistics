# Example of probablity: p(2girl/1girl)

```python
sample = 1000

families_1_girl = 0
families_2_girl = 0

for i in range(sample):
  first_child=np.random.choice(['boy','girl'])
  second_child=np.random.choice(['boy','girl'])

  if first_child == 'girl' or second_child == 'girl':
      families_1_girl += 1
      if first_child == 'girl' and second_child == 'girl':
            families_2_girl += 1
result = round(families_2_girl/families_1_girl,2)

print(f"{sample} families sampled, {families_1_girl} have one girl,{families_2_girl} have two girl")
print(result)
```

*   PERMUTATION:  n! / (n-r)!

```python
per = it.permutations([1,2,5,4,3,9],2)

for i in list(per):
  print(i)
```

* COMBINATIONS:  n! / (n-r)! * r!

same combinations dont get repeated.

```python
com = it.combinations([1,2,5,4,3,9],2)

for i in list(com):
  print(i)
```

# NORMAL DISTRIBUTION: (Continuous)

For a random variable x, with mean “μ” and standard deviation “σ”, the normal distribution formula is given by:    


f(x) = (1/√(2πσ2)) (e[-(x-μ)^2]/2σ^2).
* μ = Mean
* σ = Standard deviation
* x = Normal random variable

```python
x = np.random.normal(size=1000)    # mean value is around zero
```

```python
plt.plot(x)
plt.show()
```

```python
x.mean()
```

```python
plt.hist(x,bins=20)
plt.show()
```

* loc : mean_value (by default 0)

```python
x = np.random.normal(loc=20,size=1000)  

plt.plot(x)
plt.show()
plt.hist(x,bins=10)
plt.show()
```

```python
x = 20 + np.random.normal(size=1000)  

plt.plot(x)
plt.show()
plt.hist(x,bins=10)
plt.show()
```

STANDARD DEVIATION: "scale=" (by default 1) 

* std = 5 ---- most of the values between (+/-) 5 from mean. 
* variance = 25 (5**2) ---- left_side upto 25 & right_side upto 25 from mean.



```python
x = np.random.normal(scale=5,size=1000)  

plt.plot(x)
plt.show()
plt.hist(x,bins=20)
plt.show()
```

# Binomial distribution: (Discrete)

* A Binomial Distribution shows either (S)uccess or (F)ailure.
* used for linear regression data

```python
y = np.random.binomial(n=100,p=0.5,size=1000)  

plt.plot(y)
plt.show()
plt.hist(y,bins=20)
plt.show()
sns.displot(y)
plt.show()
```

```python
y.mean()
```

# Normal & Binomial distribution combinations:

both distribution in same graph:

* loc : mean value
* scale : std (+/- from mean value)

* n = 100 ---- 0 to 100
* p = 0.6 ---- 100*0.6 = 60 (60 is considered as Mean in binomial)

```python
x = np.random.normal(loc=60, scale=10, size=1000)
y = np.random.binomial(n=100, p=0.6,size=1000)

sns.kdeplot(x, label='normal')
sns.kdeplot(y,label='binomial')

plt.legend()
plt.plot()
```

* in code, std value should be given.
* in graph, variance is displayed.

* loc (mean) = 60 (occurance of maximum values)
* scale(standard deviation) = 5  ----  55 to 65 (occurance of more values)
* variance (square of std) = 5**2 = 25 ---- 35 to 85 (min & max) (spread of values)

```python
x = np.random.normal(loc=60, scale=5, size=1000)
y = np.random.binomial(n=100, p=0.6,size=1000)

sns.kdeplot(x, label='normal')
sns.kdeplot(y,label='binomial')

plt.legend()
plt.plot()
```
